## Train-Test Split
The dataset was divided into training and testing sets to evaluate the modelâ€™s ability to predict sales on unseen data.


```python
X = df[['TV' , 'Radio', 'Newspaper']]
y = df['Sales']
```

## Model Training
A Linear Regression model was trained to learn the relationship between advertising expenditure and sales.


```python
from sklearn.model_selection import train_test_split

X_train, X_test , y_train, y_test = train_test_split(
    X,y,test_size = 0.2
)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```
